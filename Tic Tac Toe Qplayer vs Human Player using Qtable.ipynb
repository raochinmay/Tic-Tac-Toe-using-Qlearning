```python
import numpy as np
import random
from numpy import *
import pickle
pickle_in = open("Qtable.pickle","rb")
Qtable = pickle.load(pickle_in)


def winner(player):
    win = [[[0,0],[0,1],[0,2]],[[1,0],[1,1],[1,2]],[[2,0],[2,1],[2,2]],[[0,0],[1,0],[2,0]],[[0,1],[1,1],[2,1]],[[0,2],[1,2],[2,2]],[[0,0],[1,1],[2,2]],[[0,2],[1,1],[2,0]]]
    count = 0
    for j in range(len(win)):
        count = 0
        for k in range(len(player)):
            for l in range(len(win[j])):
                if player[k] == win[j][l]:
                    count = count + 1
                    if count == 3:
                        return 1;
                        break
            else:
                continue
            break
        else:
            continue
        break
    return 0;

def getstate(availablemoves,player):
    length = len(availablemoves)
    s = ''
    for i in range(3):
        for j in range(3):
            if [i,j] in availablemoves:
                s = s + '9'
            elif [i,j] in player:
                s = s + 'X'
            else:
                s = s + 'O'
    return s

def get_move_from_key(key):
    move_key = key.split(',')[0]
    idx = move_key.index('X')
    (x,y) = divmod(idx,3)
    move = [x,y]
    return move
  
def best_move(state):
    values = Qtable[state]
    player1_move = values[0]
    max_val = double(values[0].split(',')[1])
    if len(values)>1:
        for j in range(1,len(values)):
            curr_val = double(values[j].split(',')[1])
            if curr_val > max_val:
                max_val = curr_val 
                player1_move = values[j]
    return player1_move
    
def displayboard(unoccupied,player1,player2):
    s=''
    for i in range(3):
        for j in range(3):
            if [i,j] in player1:
                s = s + 'X'
            elif [i,j] in player2:
                s = s + 'O'
            else:
                s = s + str((3*i)+j+1)
    print('\n',s[0],'|',s[1],'|',s[2],'\n----------\n',s[3],'|',s[4],'|',s[5],'\n----------\n',s[6],'|',s[7],'|',s[8])
    
def move_from_player(unoccupied):
    move_pl = input("\n Your move:  ")
    if move_pl == '':
        print('Try again')
        return [3,3]
    move_pl = int(move_pl)
    if move_pl < 1 or move_pl > 9:
        print('Invalid move')
        return [3,3]
    else:
        idx = move_pl - 1
    (x,y) = divmod(idx,3)
    #print(x,y)
    move = [x,y]
    if move not in unoccupied:
        print('Illegal move')
        return[3,3]
    return move

    
game='y'    
while game=='y' or game=='Y':
    Board = [[0,0],[0,1],[0,2],[1,0],[1,1],[1,2],[2,0],[2,1],[2,2]]
    #     print(iteration)
    player1 = [[] for z in range(5)]
    player2 = [[] for z in range(5)]
    if random.random() < 0.5:
        start_player = 'player1' 
    else:
        start_player = 'player2'

    unoccupied = Board
    #    print (unoccupied)
    for i in range(9):
        if start_player == 'player1':
            if i==0 or int(i%2)==0:
                state = getstate(unoccupied,player1)
                player1_move_key = best_move(state)    
                player1_move = get_move_from_key(player1_move_key)
                if i==0:
                    player1[i] = player1_move
                else:
                    player1[int(i/2)] = player1_move 

                unoccupied.remove(player1_move)
                displayboard(unoccupied,player1,player2)
                win1 = winner(player1)
                if win1 == 1:
                    print('Computer wins')
                    break

            elif int(i%2)==1:
                player2_move = move_from_player(unoccupied)
                while player2_move==[3,3]:
                    player2_move = move_from_player(unoccupied)
                if i==1:
                    player2[0] = player2_move
                else:
                    player2[int((i-1)/2)] = player2_move

                win2 = winner(player2)
                unoccupied.remove(player2_move)
                displayboard(unoccupied,player1,player2)
                if win2 == 1:
                    print('player wins')
                    break

        elif start_player == 'player2':
            if i==0 or int(i%2)==0:
                if i==0:
                    displayboard(unoccupied,player1,player2) 
                player2_move = move_from_player(unoccupied)
                while player2_move==[3,3]:
                    player2_move = move_from_player(unoccupied)
                if i==0:
                    player2[i] = player2_move
                else:
                    player2[int(i/2)] = player2_move

                win2 = winner(player2)
                unoccupied.remove(player2_move)
                displayboard(unoccupied,player1,player2)
                if win2 == 1:
                    print('player wins')
                    break


            elif int(i%2)==1:
                num = np.random.random()
                state = getstate(unoccupied,player1)
                player1_move_key = best_move(state)    
                player1_move = get_move_from_key(player1_move_key)
                if i==0:
                    player1[i] = player1_move
                else:
                    player1[int(i/2)] = player1_move

                win1 = winner(player1)
                unoccupied.remove(player1_move)   #loop2
                displayboard(unoccupied,player1,player2)
                if win1 == 1:
                    print('Computer wins')
                    break

    if win1!=1 and win2!=1:
        reward = 0.5
        print('cats game')
    game = input("\n Do you wanna play the next game? \n Type y or n:  ")
#print(count) 
```
